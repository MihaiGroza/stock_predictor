```python
import yfinance as yf
import pandas as pd
```

```python
tickers =["SPY", "AAPl", "MSFT"]

main_df= pd.DataFrame()

# acquiring data from yahoo
for ticker in tickers:
    

    #get data on this ticker
    tickerData = yf.Ticker(ticker)

    #get the historical prices for this ticker
    tickerDf = tickerData.history(period='1d', start='2000-1-1', end='2020-6-1')

    #see your data
    print(tickerDf.head())
    main_df["{}".format(ticker)]=(tickerDf["Close"])
    main_df["{}_vol".format(ticker)]=(tickerDf["Volume"])


main_df.head()
```

```python
main_df.to_csv('stock_data.csv') 
```
